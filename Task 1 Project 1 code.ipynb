### Importing Library

```python
import numpy as np
```

### Array Creation

```python
# 1D array
arr_1d = np.array([1, 2, 3, 4, 5])

# 2D array
arr_2d = np.array([[1, 2, 3],
                   [4, 5, 6]])

# Using built-in NumPy methods
zeros = np.zeros((2, 3))
ones = np.ones((3, 3))
range_arr = np.arange(0, 10, 2)

print(arr_1d)
print(arr_2d)
```

### Indexing and Slicing

```python
# Indexing
print(arr_1d[2])        # Access element
print(arr_2d[1, 2])     # Row 1, Column 2

# Slicing
print(arr_1d[1:4])
print(arr_2d[:, 1])     # All rows, column 1
```

### Mathematical Operations

```python
a = np.array([10, 20, 30])
b = np.array([1, 2, 3])

print(a + b)
print(a * b)
print(a ** 2)
```

### Axis-wise Operations

```python
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])

# Axis 0 → column-wise
print("Column sum:", np.sum(matrix, axis=0))

# Axis 1 → row-wise
print("Row sum:", np.sum(matrix, axis=1))
```

### Statistical Operations

```python
data = np.array([12, 15, 20, 22, 18, 30])

print("Mean:", np.mean(data))
print("Median:", np.median(data))
print(f"Standard Deviation: {np.std(data):.2f}")
print("Min:", np.min(data))
print("Max:", np.max(data))
```

### Reshaping

```python
arr = np.arange(1, 13)

reshaped = arr.reshape(3, 4)
print(reshaped)
```

### Broadcasting

```python
matrix = np.array([[1, 2, 3],
                   [4, 5, 6]])

vector = np.array([10, 20, 30])

result = matrix + vector
print(result)
```

### Save & Load NumPy Arrays

```python
# Save array
np.save("sample_array.npy", matrix)

# Load array
loaded_array = np.load("sample_array.npy")
print(loaded_array)
```

### NumPy vs Python List Performance

```python
import time

# Python list
list_data = list(range(1_000_000))
start = time.time()
list_result = [x * 2 for x in list_data]
print("List time:", time.time() - start)

# NumPy array
np_data = np.array(list_data)
start = time.time()
np_result = np_data * 2
print("NumPy time:", time.time() - start)
```
